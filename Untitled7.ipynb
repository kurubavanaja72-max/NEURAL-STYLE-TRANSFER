```python
# CODTECH TASK-4
# GENERATIVE TEXT MODEL

def generate_text(topic):
    samples = {
        "Artificial Intelligence":
        "Artificial Intelligence is transforming industries by enabling machines to perform tasks that normally require human intelligence. It is used in healthcare, education, finance, and many other fields.",

        "Machine Learning":
        "Machine Learning is a branch of AI that allows computers to learn from data and improve their performance without explicit programming.",

        "Cyber Security":
        "Cyber Security focuses on protecting computer systems, networks, and data from unauthorized access and cyber attacks."
    }

    return samples.get(
        topic,
        f"{topic} is an interesting subject with many applications and opportunities for future development."
    )

topic = input("Enter a topic: ")

print("\nGenerated Text:\n")
print(generate_text(topic))
```
